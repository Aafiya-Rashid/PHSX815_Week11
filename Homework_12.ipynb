```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


data = pd.DataFrame(np.random.randn(2000, 2), columns = ["Random_one", "Random_two"])
X = data[["Random_one", "Random_two"]]
plt.scatter(X["Random_one"], X["Random_two"], c = 'black', s = 4)
plt.xlabel('Random Numbers 1')
plt.ylabel('Random Numbers 2')


K = 4
kmeans = KMeans(n_clusters=K, random_state=0).fit(X)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

colors = ['blue', 'cyan', 'm', 'orange']
for i in range(K):
    data = X[labels == i]
    plt.scatter(data["Random_one"], data["Random_two"], c = colors[i], s = 5)

plt.scatter(centroids[:, 0], centroids[:, 1], c = 'red', s = 45, marker = "*")
plt.title('Data Points and Cluster Centroids')
plt.xlabel('Random Numbers 1')
plt.ylabel('Random Numbers 2')
plt.grid(color = 'b', alpha = 0.5, linestyle = 'dotted', linewidth = 0.7)
plt.savefig("Kmeanscluster_Random.pdf")
plt.show()






```
